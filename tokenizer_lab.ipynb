```python
from transformers import AutoTokenizer
```

```python
# Download and load the tokenizer
tokenizer = AutoTokenizer.from_pretrained('gpt2')
```

```python
print(tokenizer.vocab)
#Ġ preceding certain tokens represents a space
```

```python
#converting a sentence into tokens and IDs
tokens_id = tokenizer.encode("Here we go again! Lets learn tokenizer")

print('Tokens:  ', tokenizer.convert_ids_to_tokens(tokens_id))
print("Tokens IDs:  ", tokens_id)

#we can observe that the word "tokenizer" has been split into a combination of "token" + "izer" tokens.
```
